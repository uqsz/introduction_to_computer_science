```python
from random import randint,seed

class Node():
    def __init__(self,val):
        self.next=None
        self.val=val

def create_list(n,s):
    f=None
    seed(s)
    for i in range(1,n+1):
        p=Node(randint(0,100))
        p.next=f
        f=p
    return f

def create_sorted_list(n,s):
    f=None
    seed(s)
    l=[]
    for i in range(n):
        l.append(randint(0,100))
    l.sort(reverse=True)
    for i in range(n):
        p=Node(l[i])
        p.next=f
        f=p
    return f

def show(first):
    p=first
    while p!=None:
        print(str(p.val)+" -> ",end="")
        p=p.next
    print("")
        
f=create_list(10,12)
show(f)

f1=create_sorted_list(10,12)
show(f1)
```

```python
#3. Write a function that merges two sorted lists into one sorted list.

f1=create_sorted_list(5,11)
show(f1)
f2=create_sorted_list(5,14)
show(f2)

def merge(f1,f2):
    if f1==None and f2==None:
        return None
    if f1==None:
        return f2
    if f2==None:
        return f1
    if f1.val<f2.val:
        p1=f1
        p2=f2
        f=f1
    else:
        p1=f2
        p2=f1
        f=f2
    while p1!=None and p2!=None:
        while p1.next!=None and p1.next.val<p2.val:
            p1=p1.next
        if p1.next==None:
            p1.next=p2
            return f
        else:
            q=p1.next
            p1.next=p2
            p2=p2.next
            p1.next.next=q
    return f
            
f=merge(f1,f2)
show(f)

```
