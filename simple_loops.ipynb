```python
#2. Find the beginning words instead of 1,1 with the smallest sum, so that in the sequence analogous to the Fibonacci sequence has an expression equal to the number of the current year.

def fib(x,y,n):
    tab=[x,y]
    while y<n:
        z=x+y
        x=y
        y=z
    if y==n:
        return(tab)
    else:
        return([n,n])

def wyrazy(x,y,n):
    while x < n*n:
        print(x, end=" ")
        z = x + y
        x = y
        y = z

n=int(input())
s=n*2
y=1
x=1
while x<n:
    y=x+1
    while y<n:
        c=fib(x,y,n)[0]+fib(x,y,n)[1]
        if c<s:
            s=c
            a=fib(x,y,n)[0]
            b=fib(x,y,n)[1]
        y+=1
    x+=1
print(a,b)
wyrazy(a,b,n)
```

```python
#12. Determine the greatest common divisor of 3 given numbers.

def nwd(a,b):
    i=2
    x=max(a,b)
    y=1
    while i*i<=x:
        if a%i==0 and b%i==0:
            a/=i
            b/=i
            y*=i
            i-=1
        i+=1
    return(y)
a=int(input())
b=int(input())
c=int(input())
print(nwd(nwd(a,b),c))
```

```python
#13. Determine the smallest common multiple of 3 given numbers.

def nww(a,b):
    i=2
    x=max(a,b)
    while i*i<=x:
        if a%i==0 and b%i==0:
            a//=i
            b//=i
            i-=1
        i+=1
    return(x*min(a,b))
a=int(input())
b=int(input())
c=int(input())
print(nww(nww(a,b),c))
```

```python
#23.  Given a natural number with unique digits among which there is no zero. How many different numbers divisible by e.g. 7 can be obtained by plotting any digits in this number. For example: for 2315 it would be 21, 35, 231, 315.

import math
def zad5(n):
    d=math.floor(math.log10(n)+1)
    counter=0
    for i in range(1,2**d):
        dec=1
        a=0
        x=n
        while x!=0:
            if i%2==1:
                a+=dec*(x%10)
                dec*=10
            i//=2
            x//=10
        if a%7==0:
            counter+=1
    return counter
```
