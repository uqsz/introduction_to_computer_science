```python
#1. An array T[N][N] is given. Please write a function that fills the array with successive numbers natural numbers along a spiral

n=20
T=[[0 for i in range(n)] for j in range(n)]
a=0
x=1
while n>1:
    for j in range(a,n-1):
        T[a][j]=x
        x+=1
    for i in range(a,n-1):
        T[i][n-1]=x
        x+=1
    for j in range(n-1,a,-1):
        T[n-1][j]=x
        x+=1
    for i in range(n-1,a,-1):
        T[i][a]=x
        x+=1
    n-=1
    a+=1
if n==1:
    T[a-2][a-2]=x
for i in range(20):
    print(T[i])
```

```python
#2. An array T[N][N] filled with natural numbers is given. Please write a function that answers the question, whether in each row of the array there is at least one number consisting only of odd digits.

from random import randint
T=[[randint(0,500) for _ in range(10)] for _ in range(10)]
def nieparzysta(T):
    N=len(T)
    l=True
    for i in range(N):
        if l:
            l=False
            for j in range(N):
                a=T[i][j]
                while a>0:
                    if a%2==0:
                        break
                    a//=10
                else:
                    l=True
                    break
        else:
            return l
            break
    return l

print(nieparzysta(T))
```

```python
#4. An array T[N][N] filled with natural numbers is given. Please write a function that returns the row and column of any element for which the quotient of the sum of the elements in the column in which the element to the sum of the elements of the row in which the element lies is the largest.

from random import randint
T=[[randint(-50,50) for _ in range(3)] for _ in range(3)]
def maks(T):
    N=len(T)
    line=[0 for i in range(N)]
    col=line.copy()
    for i in range(N):
        for j in range(N):
            line[i]+=T[i][j]
            col[i]+=T[j][i]
    b=0
    for i in range(N):
        for j in range(N):
            if line[j]!=0:
                a=col[i]/line[j]
                if max(a,b)>b:
                    x=i
                    y=j
                    b=max(a,b)
    return x+1,y+1
print(maks(T))
for i in range(len(T)):
    print(T[i])
```

```python
#6. An array T[N][N] filled with natural numbers is given. Please write a function that in searches the array for the longest geometric sequence lying diagonally in the right-bottom direction, counting at least 3 elements. An array should be passed to the function. The function should return information whether it was able to found such a sequence and the length of this sequence.

# -*- coding: utf-8 -*-
from random import randint
def geo(t):
    n=len(t)
    if n<3:
        return 0
    maks=0
    for i in range(3,n+1):
        for j in range(n-i+1):
            for k in range(j+1,i+j-1):
                if t[k-1]*t[k+1]!=t[k]**2 or t[k]==0:
                    break
                if k==i+j-2:
                    maks=max(maks,i)     
    return maks
def zeruj(t):
    for i in range(len(t)):
        t[i]=0
def zad8(T):
    maks=0
    N=len(T)
    t=[0 for _ in range(N*2+1)]
    for y in range(N):
        for k in range(N-y):
            t[k]=T[y+k][k]
            t[N+k+1]=T[k][y+k]
        if maks<geo(t):
            print(t)
        maks=max(maks,geo(t))
        zeruj(t)
    if maks>2:
        return True,maks
    else:
        return False
T=[[randint(1,20) for _ in range(10)] for _ in range(10)]
for i in range(len(T)):
    print(T[i])
print(zad8(T))
```

```python
#8. Write a function that, for an array T[N][N], filled with integers, returns a value of True if there is at least one 0 in each row and each column, and the value False in the opposite case.

from random import randint
def zad10(T):
    N=len(T)
    t=[False for _ in range(N*2)]
    for i in range(N):
        for j in range(N):
            if T[i][j]==0:
                t[i]=True
                t[N+j]=True
    for i in range(N*2):
        if t[i]==False:
            return False
    return True
T=[[randint(0,2) for _ in range(5)] for _ in range(5)]
for i in range(len(T)):
    print(T[i])
print(zad10(T))
```

```python
#9. Two natural numbers are "friends if the sets of digits from which the numbers are constructed are identical. For example: 123 and 321, 211 and 122, 35 3553. Given an array T[N][N] filled with numbers natural numbers. Please write a function that, for the array T, returns how many elements of the array sit exclusively with friends.

from random import randint

def cyfry(a):
    t=[False for _ in range(10)]
    while a>0:
        t[a%10]=True
        a//=10
    return t

def friends(a,b):
    if cyfry(a)==cyfry(b):
        return True
    return False

def neigh(T2,x,y):
    if T2[x][y]:
        return False
    for i in range(9):
        if i==4:
            continue
        if not T2[x-1+i//3][y-1+i%3]:
            return False
    return True

def zad11(T):
    counter=0
    N=len(T)
    T1=[[False for _ in range(N)] for _ in range(N)]
    T2=[[True for _ in range(N+2)] for _ in range(N+2)]
    for i in range(N*N):
        for j in range(i+1,N*N):
            if friends(T[i//N][i%N],T[j//N][j%N]):
                T1[i//N][i%N],T1[j//N][j%N]=True,True
    for i in range(1,N+1):
        for j in range(1,N+1):
            T2[i][j]=T1[i-1][j-1]
    for y in range(1,N+1):
        for x in range(1,N+1):
            if neigh(T2,x,y):
                counter+=1
    for i in range(len(T2)):
        print(T2[i])
    return counter


T=[[randint(1,100) for _ in range(10)] for _ in range(10)]
for i in range(len(T)):
    print(T[i])
print(zad11(T))
```

```python
#10. An array T[N][N][N] is given. Please write a function to which we pass an array filled with numbers greater than zero. The function should return True if at all levels of the array the number of elements adjacent (within the level) to at least 6 complex numbers is the same or the value False otherwise.

def neigh(T,x,y,d):
    counter=0
    for i in range(9):
        if i==4:
            continue
        if T[x-1+i//3][y-1+i%3][d]:
            counter+=1
    return counter

def comp(a):
    if a<2:
        return False
    i=2
    while i*i<=a:
        if a%i==0:
            return True
        i+=1
    return False

def zad12(T):
    N=len(T)
    t=[[False for _ in range(N)] for _ in range(N)]
    a=True
    for d in range(N):
        T1=t.copy()
        s=0
        for i in range(N*N):
            if comp(T[i//N][i%N][d]):
                T1[i//N][i%N]=True
        for x in range(1,N-1):
            for y in range(1,N-1):
                if neigh(T1,x,y):
                    s+=1
        if a:
            k=s
            a=False
        elif s!=k:
            return False
    return True
```

```python
#11. The natural numbers a,b are complementary if their sum is a prime number. An array is given T[N][N] filled with natural numbers. Please write a function that zeros the elements that have no complementary number.
def prime(x):
    if x<2:
        return False
    i=2
    while i*i<=x:
        if x%i==0:
            return False
        i+=1
    return True

def zad13(x):
    T=x.copy()
    N=len(T)
    T1=[[False for _ in range(N)] for _ in range(N)]
    for i in range(N*N):
        for j in range(i+1,N*N):
            if prime(T[i//N][i%N]+T[j//N][j%N]):
                T1[i//N][i%N],T1[j//N][j%N]=True,True
    for i in range(N*N):
        if not T1[i//N][i%N]:
            T[i//N][i%N]=0
    return T
```

```python
#15. An array T[N][N] filled with natural numbers is given. Please write a function that returns the row and column of any element for which the sum of the surrounding elements is the largest.
from random import randint
def neigh(T,x,y):
    s=0
    for i in range(9):
        s+=T[y-1+i//3][x-i+i%3]
    s-=T[x][y]
    return s
def zad17(T):
    N=len(T)
    T1=[[0 for _ in range(N+2)] for _ in range(N+2)]
    for i in range(1,N-1):
        for j in range(1,N-1):
            T1[i][j]=T[i-1][j-1]
    maks=0
    for i in range(1,N-1):
        for j in range(1,N-1):
            x=neigh(T1,i,j)
            if x>=maks:
                maks=x
                p=i
                q=j
    return maks,p+1,q+1
T=[[randint(1,100) for _ in range(5)] for _ in range(5)]
for i in range(5):
    print(T[i])
print(zad17(T))
```

```python
#16. An array T[N][N] filled with integers is given. Please write a function that finds a consistent subsequence of elements lying horizontally or vertically with the largest sum. The maximum length of the subsequence can be 10 elements. An array T should be passed to the function, the function should return the sum of the of the maximum subsequence.

from random import randint
def sumy(T,k):
    N=len(T)
    maks=0
    for i in range(k,10+k):
        sk=0
        sw=0
        for j in range(k,10+k):
            sw+=T[i][j]
            sk+=T[j][i]
        maks=max(maks,sw,sk)
    return maks
def zad18(T):
    N=len(T)
    s=0
    if N<=10:
        return sumy(T,0)
    else:
        for k in range(N-9):
            s=0
            s=max(s,sumy(T,k))
        return s
```

```python
#17. An array T[N][N] filled with natural numbers is given. Please write a function that returns the number of pairs of elements of a certain product, such that the elements are one move away from each other chess jumper move.

from random import randint
def skoczek(T,x,y,k):
    s=0
    for a in range(-1,2,2):
        for i in range(-2,3):
            if i==0:
                continue
            if T[x+i][y+a]*T[x][y]==k:
                s+=1
            elif T[x+i][y+2*a]*T[x][y]==k:
                s+=1
            elif T[x+i][y+2*a]*T[x][y]==k:
                s+=1
            elif T[x+i][y+a]*T[x][y]==k:
                s+=1
    return s

def zad19(T,k):
    N=len(T)
    s=0
    T1=[[0 for _ in range(N+4)] for _ in range(N+4)]
    for i in range(2,N+2):
        for j in range(2,N+2):
            T1[i][j]=T[i-2][j-2]
    for i in range(2,N+2):
        for j in range(2,N+2):
            s+=skoczek(T1,i,j,k)
    return s//2
    
T=[[randint(1,5) for _ in range(5)] for _ in range(5)]
for i in range(5):
    print(T[i])
print(zad19(T,63))
```

```python
#18. Given an array T[N][N] (representing a chessboard) filled with natural numbers. Please write a function that sets two towers on the chessboard, so that the sum of the numbers on the "chessed" by the towers was the largest. To the function should be passed an array, the function should return the position of the towers. Note - we assume that the tower chesses the entire row and column, excluding the field on which it stands.

from random import randint
def zad20(T):
    N=len(T)
    maks=0
    for i in range(N*N):
        for j in range(i+1,N*N):
            s=0
            if i//N==j//N:
                for x in range(N):
                    s+=T[i//N][x]+T[N*x//N][i%N]+T[N*x//N][j%N]
            elif i%N==j%N:
                for x in range(N):
                    s+=T[x][i%N]+T[i//N][x]+T[j//N][x]
            else:
                for x in range(N):
                    s+=T[i//N][x]+T[N*x//N][i%N]+T[j//N][x]+T[N*x//N][j%N]
                s=s-T[i//N][j%N]-T[j//N][i%N]
            s=s-(T[i//N][i%N]+T[j//N][j%N])*2
            if s>maks:
                maks=max(s,maks)
                a,b,c,d=i//N,i%N,j//N,j%N
    return maks,a,b,c,d
    
T=[[randint(0,2) for _ in range(3)] for _ in range(3)]
for i in range(3):
    print(T[i])
print(zad20(T))
```
