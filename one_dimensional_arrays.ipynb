```python
#1. Write a function that converts and outputs a natural number to a system with base 2-16.

def convert(n,s):
    num=""
    x=n
    T=["A","B","C","D","E","F"]
    while x!=0:
        c=x%s
        if c<10:
            num+=str(c)
        else:
            num+=T[c-10]
        x//=s
    return num[::-1]
```

```python
#3. Write a program that generates and outputs prime numbers less than N using the Sieve of Eratosthenes method.

def primes(N):
    T=[True for _ in range(N+1)]
    for i in range(2,N+1):
        j=2
        while j*i<N+1:
            T[j*i]=False
            j+=1
    return T
```

```python
#6. Write a function that, for an N-element array t filled with natural numbers, determines the the length of the longest coherent increasing subsequence.

def zad9(t):
    maks=0
    counter=1
    for i in range(len(t)-1):
        if t[i]<t[i+1]:
            counter+=1
            maks=max(counter,maks)
        else:
            counter=1
    return maks
```

```python
#7. Write a function that, for an N-element array t filled with natural numbers, determines the the length of the longest consistent arithmetic subsequence.

def zad10(t):
    maks=0
    counter=1
    i=0
    while i<len(t)-1:
        r=t[i+1]-t[i]
        while i<len(t)-1 and r==t[i+1]-t[i]:
            i+=1
            counter+=1
            maks=max(counter,maks)
        counter=1
        i+=1
    return maks
```

```python
#8. Write a function that, for an N-element array t filled with natural numbers, determines the length of the longest consistent geometric subsequence.

def zad11(t):
    maks=0
    counter=1
    i=0
    while i<len(t)-1:
        q=t[i+1]/t[i]
        while i<len(t)-1 and q==t[i+1]/t[i]:
            i+=1
            counter+=1
            maks=max(counter,maks)
        counter=1
        i+=1
    return maks
```

```python
#9. write a program that fills an N-element array t with pseudorandom numbers from the range [1..99] and then determines and prints the difference between the length of the longest arithmetic sequence with a positive difference, and the length of the longest arithmetic sequence with a negative difference, assuming that successive expressions of the sequence are elements of the array with successive indices.

def zad12(t):
    n=len(t)
    i=0
    l1=1
    l2=1
    maks1=0
    maks2=0
    while i<n-1:
        r=t[i+1]-t[i]
        if r>0:
            while i<n-1 and t[i+1]-t[i]==r:
                l1+=1
                i+=1
            maks1=max(l1,maks1)
        elif r<0:
            while i<n-1 and t[i+1]-t[i]==r:
                l2+=1
                i+=1
            maks2=max(l2,maks2)
        if l1!=1 or l2!=1:
            i-=1
        l1=1
        l2=1
        i+=1
    return(maks1-maks2)
```

```python
#12. A large array t is given. Please write a function that returns information whether in the array the following condition is true: "all elements whose index is an element of the Fibonacci sequence are complex numbers, and among the remaining elements at least one is a prime number".

def fib(n):
    x=1
    y=1
    for _ in range(n-1):
        z=x+y
        x=y
        y=z
    return y


def prime(n):
    if n==1:
        return False
    x=n
    i=2
    while i*i<=x:
        if x%i==0:
            return False
        i+=1
    return True

def zad15(t):
    a=False
    c=len(t)-1
    j=1
    for i in range(c):
        while fib(j)<=c:
            if i==fib(j):
                if prime(t[i]):
                    return False
            j+=1
        j=1
        if not a and prime(t[i]):
            a=True
    return a
```

```python
#15. Given an N-element array t is filled with natural numbers. Please write function, which returns the length of the longest coherent subsequence being a palindrome consisting only of odd numbers. The function should return the length of the found subsequence, or 0 if there is no such subsequence.

def zad18(t):
    maks=0
    n=len(t)
    for i in range(n):
        if t[i]%2==0:
            continue
        for j in range(i,n):
            if t[j]%2==0:
                break
            a,b=i,j
            while a<b:
                if t[a]!=t[b]:
                    break
                a+=1
                b-=1
            else:
                maks=max(j-i+1,maks)
    return maks
```

```python
#17. Given an N-element array t containing natural numbers less than 1000, please write a function that returns the length of the longest coherent part of the array for which in the product of its elements each prime factor occurs at most once. For example, for the array t=[2,23,33,35,7,4,6,7,5,11,13,22] the result is a value of 5.

def wp(a,b):
    x=max(a,b)
    y=min(a,b)
    i=2
    while i*i<=y:
        if y%i==0 and x%i==0:
            return False
        i+=1
    return True

def zad20(t):
    n=len(t)
    maks=1
    for i in range(n):
        x=t[i]
        for j in range(i+1,n):
            if not wp(x,t[j]):
                break
            maks=max(maks,j-i+1)
            x*=t[j]
    return maks
```
